```python
file_id = '1VCALoCoY-i702wbS1uuzY0J2KQn6TbDL1-p8EbqzZxQ'
sheet_name = 'Sheet1'

url = f'https://docs.google.com/spreadsheets/d/{file_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}'
```

```python
import pandas as pd

df = pd.read_csv(url)
df.head(10)
```

```python
import matplotlib.pyplot as plt

# Get the top ten most frequent images
top_10 = df.sort_values(by='Frequency', ascending=False).head(10)

lables =[f"{row['Actual Image']} count:({row['Frequency']})" for _, row in top_10.iterrows()]

plt.figure(figsize=(12, 6))
plt.pie(top_10['Frequency'], labels=lables, autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Most Frequent Images')
plt.axis('equal')
plt.show()
```
