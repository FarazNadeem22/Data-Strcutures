```python
df = pd.read_csv(url)
df.head(10)
```

```python
import matplotlib.pyplot as plt

# Get the top ten most frequent images
top_10 = df.sort_values(by='Frequency', ascending=False).head(10)

lables =[f"{row['Actual Image']} count:({row['Frequency']})" for _, row in top_10.iterrows()]

plt.figure(figsize=(12, 6))
plt.pie(top_10['Frequency'], labels=lables, autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Most Frequent Images')
plt.axis('equal')
plt.show()
```

```python
file_id = '1EVqq3VeNoMA5g2mjIFniPeqYKd2U_NZ9C6ivH3QlpEE'
sheet_name = 'Sheet1'

url = f'https://docs.google.com/spreadsheets/d/{file_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}'
```

```python
df = pd.read_csv(url)
df.head(10)
```

```python
# --- Bar Chart: Correct vs Incorrect ---
plt.figure()
df["Correct"].value_counts().plot(kind='bar')
plt.title("Correct vs Incorrect Predictions")
plt.ylabel("Number of Samples")
plt.xlabel("Prediction Result")
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Pie Chart: Correct vs Incorrect
plt.figure()
df["Correct"].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title("Correct vs Incorrect Predictions (Pie Chart)")
plt.ylabel("")  # Remove y-label
plt.tight_layout()
plt.show()
```
